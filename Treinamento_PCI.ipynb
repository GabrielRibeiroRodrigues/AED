```python
import cv2
import matplotlib.pyplot as plt
import torch

# Carregar o modelo treinado
model = torch.hub.load('ultralytics/yolov5', 'custom', path='runs/train/exp4/weights/best.pt')

# Caminho para a imagem
file_path = 'carroo.png'  # Substitua pelo nome da sua imagem

# Carregar a imagem usando OpenCV
img = cv2.imread(file_path)

# Converter a imagem de BGR para RGB (OpenCV usa BGR por padr茫o)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Realizar a infer锚ncia
results = model(img_rgb)

# Exibir a imagem com as detec莽玫es
results.show()

# Se preferir salvar a imagem com as detec莽玫es
results.save('output_image.jpg')

# Mostrar a imagem usando matplotlib
plt.imshow(results.imgs[0])  # Exibe a primeira imagem resultante
plt.axis('off')  # Remove os eixos
plt.show()
```
